```python
# Create Bronze Schema
spark.sql("CREATE SCHEMA IF NOT EXISTS workspace.bronze")

# Read ALL files using wildcard *
file_path = "/Volumes/workspace/default/data/yellow_tripdata_*.csv"

df_all_raw = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(file_path)

# Save everything as ONE table in Bronze
df_all_raw.write.format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("workspace.bronze.raw_yellow_taxi")

# Display to verify it combined everything
print(f"Total rows loaded: {df_all_raw.count()}")
```
